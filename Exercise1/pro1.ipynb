```python
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
G = nx.Graph()
```

```python
G.add_node("A")
G.add_node("B")
G.add_node("C")
```

```python
G.add_edge("A","B")
G.add_edge("A","C")
```

```python
nx.draw(G,with_labels=True,node_color="blue",node_size=3000,font_size=20,font_color="white")
plt.margins(0.2)
plt.show()
```
