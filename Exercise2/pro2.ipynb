```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(["A","B","C"])
G.add_edges_from(["AB","BC"])
```

```python
nx.draw(G,with_labels=True)
```

```python
G.add_nodes_from("DE")
G.add_edge("D","E")
```

```python
nx.draw(G,with_labels=True)
```

```python
import matplotlib.pyplot as plt

degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees)
plt.show()
```
